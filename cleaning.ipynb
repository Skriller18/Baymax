```python
import re
from collections import OrderedDict

# Example input text, assuming it's read from a file or another source
with open("output\\ajyZXWu4f2A.en.vtt") as file:
    input_text = file.read()

# Regular expression to match lines with the specific pattern
pattern = r'.*<\d{2}:\d{2}:\d{2}\.\d{3}><c>.*?</c>.*'

# Remove lines matching the pattern
cleaned_text = re.sub(pattern, '', input_text, flags=re.MULTILINE).strip()

# Remove duplicate lines while preserving order
lines = cleaned_text.splitlines()
unique_lines = list(OrderedDict.fromkeys(lines))

# Join the unique lines back into a single string
cleaned_text_no_duplicates = "\n".join(unique_lines)

print(cleaned_text_no_duplicates)
```

```python
import re
from collections import OrderedDict

# Example input text
input_text = cleaned_text_no_duplicates

# Regular expression to match timestamps and formatting directives
pattern = r'\d{2}:\d{2}:\d{2}\.\d{3} --> \d{2}:\d{2}:\d{2}\.\d{3}.*?[\r\n]+|<.*?>'

# Remove timestamps and formatting directives
text_only = re.sub(pattern, '', input_text).strip()

# Split into lines, remove empty lines, and append a separator at the end of each sentence
lines = text_only.splitlines()
non_empty_lines = [line.strip() + ':' for line in lines if line.strip()]
cleaned_text_with_separators = "\n".join(non_empty_lines)

print(cleaned_text_with_separators)
```

```python
with open('context.txt', 'w') as f:
    f.write(cleaned_text_with_separators)
```
