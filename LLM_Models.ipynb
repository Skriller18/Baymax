```python
from transformers import AutoTokenizer, BertForNextSentencePrediction
import torch

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
model = BertForNextSentencePrediction.from_pretrained("google-bert/bert-base-uncased")

prompt = "In Italy, pizza served in formal settings, such as at a restaurant, is presented unsliced."
next_sentence = "The sky is blue due to the shorter wavelength of blue light."
encoding = tokenizer(prompt, next_sentence, return_tensors="pt")

outputs = model(**encoding, labels=torch.LongTensor([1]))
logits = outputs.logits
assert logits[0, 0] < logits[0, 1]  # next sentence was random
```

```python
from transformers import AutoTokenizer, BertGenerationDecoder, BertGenerationConfig
import torch

tokenizer = AutoTokenizer.from_pretrained("google/bert_for_seq_generation_L-24_bbc_encoder")
config = BertGenerationConfig.from_pretrained("google/bert_for_seq_generation_L-24_bbc_encoder")
config.is_decoder = True
model = BertGenerationDecoder.from_pretrained(
    "google/bert_for_seq_generation_L-24_bbc_encoder", config=config
)

inputs = tokenizer("Hello, my dog is cute", return_token_type_ids=False, return_tensors="pt")
outputs = model(**inputs)

prediction_logits = outputs.logits
```

```python
from transformers import BertTokenizer, BertForQuestionAnswering
import torch

# Load the BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('deepset/bert-base-cased-squad2')
model = BertForQuestionAnswering.from_pretrained('deepset/bert-base-cased-squad2')
```

```python
# Define the input text and question
text = "At Lakeside Community Hospital, Emma Johnson, a dedicated receptionist with over a decade of experience, is managing a busy Monday morning. As the first point of contact, Emma is handling multiple tasks simultaneously—answering phone calls, directing patients to the appropriate departments, and registering new arrivals. A patient, Mr. David Brown, arrives at the reception desk visibly anxious and in need of immediate assistance. He explains that he has a severe headache and blurry vision, symptoms that have worsened since the previous night. Emma quickly assesses the urgency of the situation and calmly calls for a nurse to escort Mr. Brown to the emergency department. Meanwhile, she ensures that his personal information is accurately entered into the hospital’s electronic system, which is crucial for the medical team to access his medical history. While attending to Mr. Brown, Emma also assists a couple who are at the hospital for a prenatal check-up, making sure they know where to go and what forms to fill out. Despite the high volume of patients and inquiries, Emma remains composed and efficient, embodying the critical role of hospital reception in patient care and experience."
'''with open('context.txt','r') as f:
    text = f.read()'''
question = "What tasks was Emma Johnson handling at the hospital reception?"

# Encode the input text and question, and get the scores for each word in the text
input = tokenizer(question, text,  return_tensors="pt")

output = model(**input)

# Find the words in the text that corresponds to the highest start and end scores
# with a torch.no_grad():
#     outputs = model(**inputs)
start_index = output.start_logits.argmax()
end_index = output.end_logits.argmax() + 1

# Extract the span of words as the answer
answer = tokenizer.decode(input.input_ids[0, start_index:end_index])
print(answer)
```

```python
from transformers import DistilBertTokenizer, DistilBertForQuestionAnswering

# Load the tokenizer and model for question answering in a GPT2 model
tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
model = DistilBertForQuestionAnswering.from_pretrained("distilbert-base-uncased-distilled-squad")
```

```python


# Define the input text and question
text = "At Lakeside Community Hospital, Emma Johnson, a dedicated receptionist with over a decade of experience, is managing a busy Monday morning. As the first point of contact, Emma is handling multiple tasks simultaneously—answering phone calls, directing patients to the appropriate departments, and registering new arrivals. A patient, Mr. David Brown, arrives at the reception desk visibly anxious and in need of immediate assistance. He explains that he has a severe headache and blurry vision, symptoms that have worsened since the previous night. Emma quickly assesses the urgency of the situation and calmly calls for a nurse to escort Mr. Brown to the emergency department. Meanwhile, she ensures that his personal information is accurately entered into the hospital’s electronic system, which is crucial for the medical team to access his medical history. While attending to Mr. Brown, Emma also assists a couple who are at the hospital for a prenatal check-up, making sure they know where to go and what forms to fill out. Despite the high volume of patients and inquiries, Emma remains composed and efficient, embodying the critical role of hospital reception in patient care and experience."
question = "What tasks was Emma Johnson handling at the hospital reception?"

# Encode the input text and question
inputs = tokenizer.encode_plus(question, text, add_special_tokens=True, return_tensors="pt")

# Get the model's prediction
outputs = model(**inputs)

# Find the start and end positions of the answer
start_index = outputs.start_logits.argmax()
end_index = outputs.end_logits.argmax() + 1

# Extract the answer
answer_tokens = inputs.input_ids[0][start_index:end_index]
answer = tokenizer.decode(answer_tokens)
print(answer)
```
